```python
import numpy as np
import pandas as pd
import seaborn as sn
from matplotlib import pyplot as plt
```

```python
diamonds = pd.read_csv('csv/diamonds.csv')
```

```python
diamonds.head(10)
```

```python
diamonds['cut']
```

```python
diamonds.duplicated().value_counts()
```

```python
diamonds["cut"] = diamonds["cut"].str.lower()
```

```python
cuts = tuple(set(diamonds["cut"]))
index = np.arange(len(cuts))

plt.hist(diamonds["cut"], color="red", bins=len(cuts))
plt.title('Cut Histogram')
plt.xlabel("\nCut")
plt.ylabel("Frequency\n")
plt.xticks(index, cuts)
plt.show()
```

```python
diamonds.corr()
```

```python
plt.matshow(diamonds.corr())
index = np.arange(len(diamonds.columns))
plt.xticks(index, diamonds.columns)
plt.yticks(index, diamonds.columns)
plt.colorbar()
plt.show()
```
